## Choices analysis

Transform wide data into long data

```python
sample_df[['ResponseId']+choice_Qs].drop([0,1]).head(3)
```

```python
# Transform the wide data to long data.
choices_df = pd.melt(
    sample_df[['ResponseId']+choice_Qs].drop([0,1]), 
    id_vars=['ResponseId'], value_vars=choice_Qs,
    var_name='QID', value_name='choice'
).dropna()
print('%s choice responses' % len(choices_df))
choices_df.head(3)    
```

Parse out the choice parameters from the Question ID

```python
choices_df['choice'] = choices_df['choice'].astype(int)
choices_df['v'] =  choices_df['QID'].apply(lambda QID: int(QID.split('_v')[1][0]))
choices_df['gc'] =  choices_df['QID'].apply(lambda QID: int(QID.split('_gc')[1][0]))
choices_df['gt'] =  choices_df['QID'].apply(lambda QID: int(QID.split('_gt')[1][0]))
choices_df['dc'] =  choices_df['QID'].apply(lambda QID: int(QID.split('_dc')[1][0]))
# oops! early version of the question generation script coded what should be dt as dc
# in the QID
# the values in the questions were still the dt_value though.
# there is no good way to fix this data. Will just have to assume same value as gt for now
choices_df['dt'] =  choices_df['gt']
choices_df['dp'] =  choices_df['QID'].apply(lambda QID: int(QID.split('_dp')[1][0]))

choices_df.head(3)
```

Add values to the question parameters

```python
import sys  
sys.path.insert(0, '../survey-questions')

from attributes import (
    VENDOR_TYPES_V_ICRAT, COST_LEVELS_V_ICRAT, 
    TIME_LEVELS_V_ICRAT, DRONE_PRIVACY_LEVELS_V_ICRAT,
)
```

```python
choices_df['v_type'] = choices_df['v'].apply(lambda v: VENDOR_TYPES_V_ICRAT[v])
choices_df['g_cost'] = choices_df.apply(lambda row: COST_LEVELS_V_ICRAT[row['v']][row['gc']],axis=1)
choices_df['d_cost'] = choices_df.apply(lambda row: COST_LEVELS_V_ICRAT[row['v']][row['dc']],axis=1)
choices_df['g_time'] = choices_df.apply(lambda row: TIME_LEVELS_V_ICRAT[row['v']][row['gt']],axis=1)
choices_df['d_time'] = choices_df.apply(lambda row: TIME_LEVELS_V_ICRAT[row['v']][row['dt']],axis=1)
# choices_df['g_time'] = choices_df.apply(lambda row: TIME_LEVELS_V_ICRAT[v][t])

choices_df.head()
```

### Note

Due to the data error, we just compare cost.
We pretend wait times are always the same, because we cannot recover the true drone delivery wait time from the Question ID. This is not a valid assumption!

```python
# ($) drone delivery fee - ground vehicle delivery fee
choices_df['d_cost - g_cost'] = choices_df['d_cost'] - choices_df['g_cost']
choices_df.head()
```

```python
fig,ax = plt.subplots(1,1,figsize=(5,3),dpi=100)
(choices_df
 .rename(columns={'dp':'drone privacy'})
 .groupby(['d_cost - g_cost','drone privacy'])['choice'].mean()
 .unstack()
 .rename(columns={0:'NO privacy', 1:'YES privacy'})
 .plot(kind='bar',ax=ax)
)
ax.set_ylabel('fraction chose drone')
_ = ax.set_xlabel('drone delivery fee - ground vehicle delivery fee')
```
